Statische KNN Klassifizierung

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import f1_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
import time

#safe measures in list
accuracy = []
f1 = []
precision = []
recall = []
kappa = []
pred_time = []
```

```python
#load data
agr_g = pd.read_csv('https://raw.githubusercontent.com/scikit-multiflow/streaming-datasets/master/agr_g.csv')

#divide features and target class
X = agr_g.drop(agr_g.iloc[:,-1:], axis=1)
y = agr_g.iloc[:,-1:]

#set test_size
n = 10000/len(agr_g)

#split test and train data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = n)

#set classifier
knn = KNeighborsClassifier(n_neighbors=3)

# fit model with train data
knn.fit(X_train, y_train.values.ravel())

#classify test data
start = time.time()
y_pred = knn.predict(X_test)
stop = time.time()

#safe measures
accuracy.append(accuracy_score(y_test, y_pred))
f1.append(f1_score(y_test, y_pred, average='weighted'))
precision.append(precision_score(y_test, y_pred, average='weighted'))
recall.append(recall_score(y_test, y_pred, average='weighted'))
kappa.append(cohen_kappa_score(y_test, y_pred))
pred_time.append(stop-start)
```
