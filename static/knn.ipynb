Statische KNN Klassifizierung

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import f1_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
import time

#set test size
p = 1000

#safe measures in list
accuracy = []
f1 = []
precision = []
recall = []
kappa = []
train_time = []
pred_time = []

#evaluate + safe measures in lists
def evaluate():
    accuracy.append(accuracy_score(y_test, y_pred))
    f1.append(f1_score(y_test, y_pred, average='weighted'))
    precision.append(precision_score(y_test, y_pred, average='weighted', zero_division=0)) # zero_divison wird gesetzt damit ohne warnung 0 ausgegeben wird weil true positive + false positive == 0 
    recall.append(recall_score(y_test, y_pred, average='weighted'))
    kappa.append(cohen_kappa_score(y_test, y_pred))
    train_time.append(stop_train-start_train)
    pred_time.append(stop_pred-start_pred)
```

```python
#load data
led_g1 = pd.read_csv('data/led_g1.csv')

#divide features and target class
X = led_g1.drop(led_g1.iloc[:,-1:], axis=1)
y = led_g1.iloc[:,-1:]

#set test_size
n = p/len(led_g1)

#split test and train data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = n)

#set classifier
knn = KNeighborsClassifier(n_neighbors=10)

# fit model with train data
start_train = time.time()
knn.fit(X_train, y_train.values.ravel())
stop_train = time.time() 

#classify test data
start_pred = time.time()
y_pred = knn.predict(X_test)
stop_pred = time.time()

#evaluate prediction
evaluate()
```
