```python
from skmultiflow.data import ConceptDriftStream
from skmultiflow.data.sea_generator import SEAGenerator
from skmultiflow.data.led_generator_drift import LEDGeneratorDrift
import numpy as np
import pandas as pd
```

```python
sea_g2_stream = ConceptDriftStream(stream=SEAGenerator
                                   (classification_function=1, 
                                    random_state=None, balance_classes=False, 
                                    noise_percentage=0.39), 
                   drift_stream=SEAGenerator(classification_function=3, 
                                    random_state=None, balance_classes=False, 
                                    noise_percentage=0.39),
                                   position=5000, width=1000, random_state=None)

stream = sea_g2_stream.next_sample(100000)
sea_g2 = pd.DataFrame(np.column_stack(stream),columns=['att_num_0', 'att_num_1', 'att_num_2', 'target_0'])
sea_g2.to_csv('sea_g2.csv')
```

```python
sea_g3_stream = ConceptDriftStream(stream=SEAGenerator
                                   (classification_function=1, 
                                    random_state=None, balance_classes=False, 
                                    noise_percentage=0.59), 
                   drift_stream=SEAGenerator(classification_function=3, 
                                    random_state=None, balance_classes=False, 
                                    noise_percentage=0.59),
                                   position=5000, width=1000, random_state=None)

stream = sea_g3_stream.next_sample(100000)
sea_g3 = pd.DataFrame(np.column_stack(stream),columns=['att_num_0', 'att_num_1', 'att_num_2', 'target_0'])
sea_g3.to_csv('sea_g3.csv')
```

```python
led_g2_stream = LEDGeneratorDrift(random_state = 112, noise_percentage = 0.35,has_noise= True,
n_drift_features=2) #es werden 2 Features mit irrelevanten getauscht

stream = led_g2_stream.next_sample(100000)
led_g2_stream = pd.DataFrame(np.column_stack(stream))
led_g2_stream.to_csv('led_g2.csv')
```
