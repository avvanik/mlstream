In diesem Notebook werden die in impl(1) generierten Streams um einen Zeitindex erweitert und in einem statischen Datensatz gespeichert. Die Streams werden auf eine Länge von 10.000 Instanzen verkürzt.

```python
import numpy as np
import pandas as pd

from skmultiflow.data import ConceptDriftStream
from skmultiflow.data.agrawal_generator import AGRAWALGenerator
from skmultiflow.data.sea_generator import SEAGenerator
from skmultiflow.data.led_generator_drift import LEDGeneratorDrift
```

AGRAWAL Generator -> t_agr_a

1. 10.000 Instanzen
2. 19 abrupte Drifts

```python
import warnings
warnings.filterwarnings('ignore')

agr_a_stream1 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=0, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=2, 
                                    balance_classes=False),
                                   position=500, width=1) 

agr_a_stream2 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=1, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=3, 
                                    balance_classes=False),
                                   position=500, width=1)

agr_a_stream3 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=4, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=6, 
                                    balance_classes=False),
                                   position=500, width=1)

agr_a_stream4 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=5, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=7, 
                                    balance_classes=False),
                                   position=500, width=1)

agr_a_stream5 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=9, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=8, 
                                    balance_classes=False),
                                   position=500, width=1)

agr_a = pd.DataFrame(np.column_stack(agr_a_stream1.next_sample(1000)))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream2.next_sample(1000))))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream3.next_sample(1000))))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream4.next_sample(1000))))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream5.next_sample(1000))))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream1.next_sample(1000))))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream2.next_sample(1000))))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream3.next_sample(1000))))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream4.next_sample(1000))))
agr_a = agr_a.append(pd.DataFrame(np.column_stack(agr_a_stream5.next_sample(1000))))

agr_a.columns = ['salary', 'comission', 'age', 'elevel',
                 'car', 'zipcode', 'hvalue', 'hyears', 'loan', 'class']

agr_a = agr_a.astype({'age': 'int', 'elevel': 'int', 
                      'car': 'int', 'zipcode': 'int',
                      'hyears': 'int', 'class': 'int'})

timestamp = pd.date_range(start=pd.Timestamp('00:00:00'), end=pd.Timestamp('00:01:40'), freq='L')
#timestamp to df
timestamp = pd.DataFrame(timestamp, columns = ['datetime'])
#delete last row
timestamp.drop(timestamp.tail(1).index, inplace=True)

agr_a.reset_index(drop=True, inplace=True)

agr_a['datetime'] = timestamp

agr_a.to_csv('t_agr_a.csv')
agr_a
```

AGRAWAL Generator -> t_agr_g1

1. 10.000 Instanzen
2. 4 graduelle Drifts (Breite: 500)

```python
import warnings
warnings.filterwarnings('ignore')

agr_g1_stream1 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=2, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=4, 
                                    balance_classes=False),
                                   position=1500, width=500) 

agr_g1_stream2 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=4, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=6, 
                                    balance_classes=False),
                                   position=1500, width=500)

agr_g1_stream3 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=6, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=3, 
                                    balance_classes=False),
                                   position=1500, width=500)

agr_g1_stream4 = ConceptDriftStream(stream=AGRAWALGenerator
                                   (classification_function=3, 
                                    balance_classes=False), 
                   drift_stream=AGRAWALGenerator(classification_function=1, 
                                    balance_classes=False),
                                   position=1500, width=500)

agr_g1 = pd.DataFrame(np.column_stack(agr_g1_stream1.next_sample(2500)))
agr_g1 = agr_g1.append(pd.DataFrame(np.column_stack(agr_g1_stream2.next_sample(2500))))
agr_g1 = agr_g1.append(pd.DataFrame(np.column_stack(agr_g1_stream3.next_sample(2500))))
agr_g1 = agr_g1.append(pd.DataFrame(np.column_stack(agr_g1_stream4.next_sample(2500))))

agr_g1.columns = ['salary', 'comission', 'age', 'elevel',
                 'car', 'zipcode', 'hvalue', 'hyears', 'loan', 'class']

agr_g1 = agr_g1.astype({'age': 'int', 'elevel': 'int', 
                      'car': 'int', 'zipcode': 'int',
                      'hyears': 'int', 'class': 'int'})

agr_g1.reset_index(drop=True, inplace=True)

timestamp = pd.date_range(start=pd.Timestamp('00:00:00'), end=pd.Timestamp('02:46:40'), freq='S')
timestamp = pd.DataFrame(timestamp, columns = ['datetime'])

agr_g1['datetime'] = timestamp
agr_g1.to_csv('t_agr_g1.csv')
agr_g1
```

SEA Generator -> t_sea_a

1. 10.000 Instanzen
2. 19 abrupte Drifts (Drifts wiederholen sich, da die Klassifizierungsfunktionen im Stream erneut auftauchen)

```python
import warnings
warnings.filterwarnings('ignore')

sea_a_stream1 = ConceptDriftStream(stream=SEAGenerator
                                   (classification_function=1, 
                                    balance_classes=False), 
                   drift_stream=SEAGenerator(classification_function=3, 
                                    balance_classes=False),
                                   position=500, width=1) 

sea_a_stream2 = ConceptDriftStream(stream=SEAGenerator
                                   (classification_function=2, 
                                    balance_classes=False), 
                   drift_stream=SEAGenerator(classification_function=0, 
                                    balance_classes=False),
                                   position=500, width=1)



sea_a = pd.DataFrame(np.column_stack(sea_a_stream1.next_sample(1000)))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream2.next_sample(1000))))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream1.next_sample(1000))))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream2.next_sample(1000))))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream1.next_sample(1000))))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream2.next_sample(1000))))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream1.next_sample(1000))))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream2.next_sample(1000))))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream1.next_sample(1000))))
sea_a = sea_a.append(pd.DataFrame(np.column_stack(sea_a_stream2.next_sample(1000))))

sea_a.columns = ['att_num_0', 'att_num_1', 'att_num_2', 'target_0']

sea_a = sea_a.astype({'target_0': 'int'})

sea_a.reset_index(drop=True, inplace=True)

timestamp = pd.date_range(start=pd.Timestamp('00:00:00'), end=pd.Timestamp('02:46:40'), freq='S')
timestamp = pd.DataFrame(timestamp, columns = ['datetime'])

sea_a['datetime'] = timestamp

sea_a.to_csv('t_sea_a.csv')
sea_a
```

SEA Generator -> t_sea_g2

1. 10.000 Instanzen
2. 2 graduellen Drifts (Breite: 1000)
3. Rauschen 25%

```python
import warnings
warnings.filterwarnings('ignore')

sea_g2_stream1 = ConceptDriftStream(stream=SEAGenerator
                                   (classification_function=1, 
                                    balance_classes=False, noise_percentage=0.25), 
                   drift_stream=SEAGenerator(classification_function=3, 
                                    balance_classes=False, noise_percentage=0.25),
                                   position=2500, width=1000) 

sea_g2_stream2 = ConceptDriftStream(stream=SEAGenerator
                                   (classification_function=3, 
                                    balance_classes=False, noise_percentage=0.25), 
                   drift_stream=SEAGenerator(classification_function=0, 
                                    balance_classes=False, noise_percentage=0.25),
                                   position=2500, width=1000)

sea_g2 = pd.DataFrame(np.column_stack(sea_g2_stream1.next_sample(5000)))
sea_g2 = sea_g2.append(pd.DataFrame(np.column_stack(sea_g2_stream2.next_sample(5000))))

sea_g2.columns = ['att_num_0', 'att_num_1', 'att_num_2', 'target_0']

sea_g2 = sea_g2.astype({'target_0': 'int'})

sea_g2.reset_index(drop=True, inplace=True)

timestamp = pd.date_range(start=pd.Timestamp('00:00:00'), end=pd.Timestamp('02:46:40'), freq='S')
timestamp = pd.DataFrame(timestamp, columns = ['datetime'])

sea_g2['datetime'] = timestamp

sea_g2.to_csv('t_sea_g2.csv')
sea_g2
```
