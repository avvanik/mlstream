```python
import pandas as pd

df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/poker/poker-hand-testing.data")

#delete n rows
n = 800000
df = df.iloc[:-n]
df
```

```python
#create timestamp
timestamp = pd.date_range(start=pd.Timestamp('00:00:00'), end=pd.Timestamp('00:03:26'), freq='L') #millisecond
#create df from timestamp
timestamp = pd.DataFrame(timestamp, columns = ['datetime'])
timestamp
```

```python
#delete n rows
n = 206001-199999
timestamp.drop(timestamp.tail(n).index, inplace=True)
df['datetime'] = timestamp
df
```

```python
df.to_csv('pok_t.csv')
```
