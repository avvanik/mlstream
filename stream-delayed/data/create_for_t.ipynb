```python
import pandas as pd

df = pd.read_csv("covtype.csv")

#delete n rows
n = 300000
df = df.iloc[:-n]
df
```

```python
#create timestamp
timestamp = pd.date_range(start=pd.Timestamp('00:00:00'), end=pd.Timestamp('00:04:46'), freq='L') #millisecond
#create df from timestamp
timestamp = pd.DataFrame(timestamp, columns = ['datetime'])
timestamp
```

```python
#delete n rows 286.001-281.011
timestamp.drop(timestamp.tail(4990).index, inplace=True)
df['datetime'] = timestamp
df
```

```python
df.to_csv('for_t.csv')
```
