In diesem Notebook wird der Airline Datensatz verkleinert und anschließend um einen timestamp column erweitert. Damit soll simuliert werden, dass jede Instanz mit einem Abstand von einer Sekunde eintrifft.

```python
import pandas as pd

#load df
df = pd.read_csv("https://raw.githubusercontent.com/scikit-multiflow/streaming-datasets/master/airlines.csv")

#delete n rows
n = 500000
df = df.iloc[:-n]
df
```

```python
#create timestamp
timestamp = pd.date_range(start=pd.Timestamp('00:00:00'), end=pd.Timestamp('10:57:00'), freq='S')
#create df from timestamp
timestamp = pd.DataFrame(timestamp, columns = ['datetime'])
timestamp
```

```python
#delete n rows 39421-39383
timestamp.drop(timestamp.tail(38).index, inplace=True)
df['datetime'] = timestamp
df
```

```python
df.to_csv('air_t.csv')
```
